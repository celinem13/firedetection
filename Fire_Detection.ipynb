```python
import numpy as np
import os
import cv2
import matplotlib as plt
from google.colab.patches import cv2_imshow

img_path = './sample_data/805.png'
img2_path = './sample_data/image3.jpg'
img3_path = './sample_data/vb.jpeg'

imggray = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
img2gray = cv2.imread(img2_path, cv2.IMREAD_GRAYSCALE)
img3gray = cv2.imread(img3_path, cv2.IMREAD_GRAYSCALE)


img = cv2.imread(img_path)
img2 = cv2.imread(img2_path)
img3 = cv2.imread(img3_path)


hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
hsv2 = cv2.cvtColor(img2, cv2.COLOR_BGR2HSV)
hsv3 = cv2.cvtColor(img3, cv2.COLOR_BGR2HSV)

mask = cv2.inRange(hsv,(0, 125, 30), (30, 255, 255) )
mask2 = cv2.inRange(hsv2,(0, 125, 30), (30, 255, 255) )
mask3 = cv2.inRange(hsv3,(0, 125, 30), (25, 255, 255) )

cv2_imshow(mask)
cv2_imshow(img)
cv2_imshow(imggray)
```

```python
alpha=0.4
beta = (1.0 - alpha)
dst = cv2.addWeighted(mask, alpha, imggray, beta, 0.0)
cv2_imshow(dst)
dst2 = cv2.addWeighted(mask2, alpha, img2gray, beta, 0.0)
cv2_imshow(dst2)
dst3 = cv2.addWeighted(mask3, alpha, img3gray, beta, 0.0)
cv2_imshow(dst3)
```
